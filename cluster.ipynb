```python
import numpy as np
import pandas as pd
from google.colab import files

# Upload the file
uploaded = files.upload()

# Load the uploaded CSV into a dataframe
file_name = next(iter(uploaded))  # Get the first file name
df = pd.read_csv(file_name)

# Display the first few rows of the dataframe
print("Original dataset:")
print(df.head())

# Assuming 'Interested Domain' is the cluster variable
# Parameters
num_cluster_to_sample = 3  # Number of clusters to sample
sample_size_per_cluster = 5  # Number of samples to take from each cluster

# Perform cluster sampling
def cluster_sampling(df, cluster_column, num_cluster_to_sample, sample_size_per_cluster):
    # Step 1: Randomly select clusters
    unique_clusters = df[cluster_column].unique()  # Find all unique clusters
    sample_clusters = np.random.choice(unique_clusters, size=num_cluster_to_sample, replace=False)  # Select clusters

    # Step 2: Sample elements within the selected clusters
    sampled_df = pd.DataFrame()  # Initialize an empty DataFrame for sampled data
    for cluster in sample_clusters:
        cluster_data = df[df[cluster_column] == cluster]  # Data for the current cluster
        # Ensure sample size does not exceed the available elements in the cluster
        sample_size = min(len(cluster_data), sample_size_per_cluster)
        sampled_elements = cluster_data.sample(n=sample_size)  # Sample from the cluster
        sampled_df = pd.concat([sampled_df, sampled_elements])  # Concatenate to the final sampled DataFrame

    return sampled_df

# Print the actual column names in the DataFrame to identify the correct cluster column
print("Column names in the DataFrame:", df.columns)

# Replace 'Interested Domain' with the correct column name based on the output above
# For example, if the correct column name is 'Interested Domain', use:
sampled_df = cluster_sampling(df, 'Interested Domain', num_cluster_to_sample, sample_size_per_cluster)

# Display the sampled dataset
print("Sampled dataset:")
print(sampled_df)
```
